# Matplotlib 한글폰트 사용하기
## 1. 필요한 패키지 가져오기

```python
# 그래프를 노트북 안에 그리기 위해 설정
%matplotlib inline

# 필요한 패키지와 라이브러리를 가져옴
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처 (matplotlib 3.4.0 기준 일부 폰트에서 여전히 깨짐 현상 발생)
mpl.rcParams['axes.unicode_minus'] = False
```

## 2. 그래프를 그리기 위한 임의의 데이터를 만들어주기

```python
import numpy as np

data = np.random.randint(-100, 100, 50).cumsum()
data
```

## 3. 기본 폰트로 지정되었기 때문에 한글이 깨지는 상태

```python
plt.plot(range(50), data, 'r')
plt.title('시간별 간격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
```

## 7. 나눔고딕 사용을 위해 'Nanum'이 들어간 폰트만 가져오기
- 폰트를 설치했는데 원하는 폰트명을 못 가져온다면, 터미널을 열어 `mpl.get_cachedir()`로 찍히는 캐시위치로 이동해서 캐시파일을 열어본다.
- 캐시파일에 원하는 폰트리스트가 없으면 주피터노트북 혹은 콘다를 재실행한다.

```python
[(f.name, f.fname) for f in fm.fontManager.ttflist if 'Nanum' in f.name]
```

## 9. 폰트를 사용하는 방법은 3가지가 있습니다.
1. `FontProperties` 사용하기: 그래프에서 폰트가 필요한 항목마다 지정해야 함
2. `matplotlib.rcParams[]`로 전역글꼴 설정: 그래프에 설정하면 폰트가 필요한 항목에 적용됨
3. 2번의 방법을 `mpl.matplotlib_fname()`로 읽어지는 설정 파일에 직접 적어주기 (모든 노트북에 적용되며, 노트북을 열 때마다 지정할 필요가 없어 편리함)

### 1. `FontProperties` 사용하기
- 텍스트를 지정하는 항목에 사용한다. 지정한 항목에만 해당 폰트가 적용된다.
- `matplotlib.pyplot`
    - `title()`
    - `xlabel()`
    - `ylabel()`
    - `legend()`
    - `text()`
- `matplotlib.axes`
    - `set_title()`

```python
# fname 옵션을 사용하는 방법
path = '/Users/heekyjung/Library/Fonts/NanumBarunpenR.ttf'
fontprop = fm.FontProperties(fname=path, size=18)

plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.ylabel('주식 가격', fontproperties=fontprop)
plt.xlabel('시간(분)', fontproperties=fontprop)
plt.show()
```

### 2. `matplotlib.rcParams[]`로 전역글꼴 설정

```python
# 기본 설정 읽기
import matplotlib.pyplot as plt

# size, family
print('# 설정 되어있는 폰트 사이즈')
print (plt.rcParams['font.size'] ) 
print('# 설정 되어있는 폰트 글꼴')
print (plt.rcParams['font.family'] )
```

```python
# serif, sans-serif, monospace
print('serif 세리프가 있는 폰트--------')
print (plt.rcParams['font.serif']) 
print('sans-serif 세리프가 없는 폰트--------')
print (plt.rcParams['font.sans-serif']) 
print('monospace 고정폭 글꼴--------')
print (plt.rcParams['font.monospace'])
```

```python
plt.rcParams['font.family'] = 'Nanum Brush Script'
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (14,4)
```

```python
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.style.use('seaborn-pastel')
plt.show()
```

#### `rcParams` 대신 `FontProperties`와 `plt.rc`를 사용하는 방법

```python
path = '/Users/heekyjung/Library/Fonts/NanumBarunGothic.ttf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
print(font_name)
plt.rc('font', family=font_name)

fig, ax = plt.subplots()
ax.plot(data)
ax.set_title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.style.use('ggplot')
plt.show()
```
